```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import scipy
```

```python
def raw_data_vis(pid):
    print('Raw data of PID:',pid)

    output_path = "/content/drive/MyDrive/data_time_label/"+pid+"/"

    for f in os.listdir(output_path):
        if f.endswith("EDA.csv"):
            eda_path = f
        elif f.endswith("TEMP.csv"):
            temp_path = f
        elif f.endswith("BVP.csv"):
            bvp_path = f
        elif f.endswith("HR.csv"):
            hr_path = f
        elif f.endswith("ACC.csv"):
            acc_path = f

    eda_data = pd.read_csv(output_path + eda_path)
    temp_data = pd.read_csv(output_path + temp_path)
    bvp_data = pd.read_csv(output_path + bvp_path)
    hr_data = pd.read_csv(output_path + hr_path)
    acc_data = pd.read_csv(output_path + acc_path)

    time_eda_raw = np.linspace(0, 21, (eda_data['eda'].shape[0]))
    plt.figure(figsize=(13,2))
    plt.plot(time_eda_raw, eda_data['eda'], color='red')
    plt.title('Raw signals from EDA sensors', fontsize = 20)
    plt.ylabel('EDA (µs)')
    plt.xlabel('time (min)')
    plt.show()

    time_bvp_raw = np.linspace(0, 21, (bvp_data['bvp'].shape[0]))
    plt.figure(figsize=(13,2))
    plt.plot(time_bvp_raw, bvp_data['bvp'], color='green')
    plt.title('Raw signals from BVP sensors', fontsize = 20)
    plt.ylabel('BVP')
    plt.xlabel('time (min)')
    plt.show()

    time_temp = np.linspace(0, 21, (temp_data['temp'].shape[0]))
    plt.figure(figsize=(13,2))
    plt.plot(time_temp, temp_data['temp'], color='orange')
    plt.title('Raw signals from TEMP sensor', fontsize = 20)
    plt.ylabel('TEMP (°C)')
    plt.xlabel('time (min)')
    plt.show()

    time_hr = np.linspace(0, 21, (hr_data['bpm'].shape[0]))
    plt.figure(figsize=(13,2))
    plt.plot(time_hr, hr_data['bpm'], color='blue')
    plt.title('Raw signals from HR sensor', fontsize = 20)
    plt.ylabel('HR (bpm)')
    plt.xlabel('time (min)')
    plt.show()

    time_acc_x = np.linspace(0, 21, (acc_data['x'].shape[0]))
    plt.figure(figsize=(13,2))
    plt.plot(time_acc_x, acc_data['x'], color='#e2c19d')
    plt.title('Raw signals from ACC (x) sensor', fontsize = 20)
    plt.ylabel('ACC (x)')
    plt.xlabel('time (min)')
    plt.show()

    time_acc_y = np.linspace(0, 21, (acc_data['y'].shape[0]))
    plt.figure(figsize=(13,2))
    plt.plot(time_acc_y, acc_data['y'], color='#20b2aa')
    plt.title('Raw signals from ACC (y) sensor', fontsize = 20)
    plt.ylabel('ACC (y)')
    plt.xlabel('time (min)')
    plt.show()

    time_acc_z = np.linspace(0, 21, (acc_data['z'].shape[0]))
    plt.figure(figsize=(13,2))
    plt.plot(time_acc_z, acc_data['z'], color='#770088')
    plt.title('Raw signals from ACC (z) sensor', fontsize = 20)
    plt.ylabel('ACC (z)')
    plt.xlabel('time (min)')
    plt.show()
```

```python
def get_only_stress_data(file):
    stress_data = pd.read_csv(file)
    stress_data = stress_data.drop(['time_series'], axis=1)
    stress_data = stress_data[(stress_data.label != 4)]
    return stress_data
```

```python
def raw_stress_data(pid):
    print('Raw stress data of PID:',pid)

    output_path = "/content/drive/MyDrive/data_time_label/"+pid+"/"

    for f in os.listdir(output_path):
        if f.endswith("EDA.csv"):
            eda_path = f
        elif f.endswith("TEMP.csv"):
            temp_path = f
        elif f.endswith("BVP.csv"):
            bvp_path = f
        elif f.endswith("HR.csv"):
            hr_path = f

    eda_data = get_only_stress_data(output_path + eda_path)
    temp_data = get_only_stress_data(output_path + temp_path)
    bvp_data = get_only_stress_data(output_path + bvp_path)
    hr_data = get_only_stress_data(output_path + hr_path)

    time_eda_raw = np.linspace(0, 16, (eda_data['eda'].shape[0]))
    plt.figure(figsize=(13,2))
    plt.axvline(x=5, color='g', ls='--')
    plt.axvline(x=10, color='g', ls='--')
    plt.plot(time_eda_raw, eda_data['eda'], color='red')
    plt.title('Raw signals from EDA sensors', fontsize = 20)
    plt.ylabel('EDA (µs)')
    plt.xlabel('time (min)')
    plt.show()

    time_bvp_raw = np.linspace(0, 16, (bvp_data['bvp'].shape[0]))
    plt.figure(figsize=(13,2))
    plt.axvline(x=5, color='g', ls='--')
    plt.axvline(x=10, color='g', ls='--')
    plt.plot(time_bvp_raw, bvp_data['bvp'], color='green')
    plt.title('Raw signals from BVP sensors', fontsize = 20)
    plt.ylabel('BVP')
    plt.xlabel('time (min)')
    plt.show()

    time_temp = np.linspace(0, 16, (temp_data['temp'].shape[0]))
    plt.figure(figsize=(13,2))
    plt.axvline(x=5, color='g', ls='--')
    plt.axvline(x=10, color='g', ls='--')
    plt.plot(time_temp, temp_data['temp'], color='orange')
    plt.title('Raw signals from TEMP sensor', fontsize = 20)
    plt.ylabel('TEMP (°C)')
    plt.xlabel('time (min)')
    plt.show()

    time_hr = np.linspace(0, 16, (hr_data['bpm'].shape[0]))
    plt.figure(figsize=(13,2))
    plt.axvline(x=5, color='g', ls='--')
    plt.axvline(x=10, color='g', ls='--')
    plt.plot(time_hr, hr_data['bpm'], color='blue')
    plt.title('Raw signals from HR sensor', fontsize = 20)
    plt.ylabel('HR (bpm)')
    plt.xlabel('time (min)')
    plt.show()
```

# **Raw EDA signals of p04, p05, p06, p07**

```python
def get_eda(pid):

    output_path = "/content/drive/MyDrive/data_time_label/"+pid+"/"

    for f in os.listdir(output_path):
        if f.endswith("EDA.csv"):
            eda_path = f

    eda_data = get_only_stress_data(output_path + eda_path)

    return {'eda':eda_data['eda']}
```

```python
p4_5_6_7 = ["p" + str(i).zfill(2) for i in range(0, 25) if i in [4,5,6,7]]
all_data = [get_eda(pid) for pid in p4_5_6_7]
```

```python
eda0 = scipy.signal.resample(all_data[0]['eda'], len(all_data[3]['eda']))
eda1 = scipy.signal.resample(all_data[1]['eda'], len(all_data[3]['eda']))
eda2 = scipy.signal.resample(all_data[2]['eda'], len(all_data[3]['eda']))
```

```python
time_eda_raw = np.linspace(0, 15, (all_data[3]['eda'].shape[0]))
plt.figure(figsize=(12,8))

plt.axvline(x=5, color='#101010', ls='--')
plt.axvline(x=10, color='#101010', ls='--')
plt.plot(time_eda_raw, eda0, color='#ff4040', label="participant 4")
plt.plot(time_eda_raw, eda1, color='#0a75ad', label="participant 5")
plt.plot(time_eda_raw, eda2, color='#8a2be2', label="participant 6")
plt.plot(time_eda_raw, all_data[3]['eda'], color='#daa520', label="participant 7")

plt.title("Physiological responses from the four participants")
plt.ylabel('EDA (µs)')
plt.xlabel('\ntime (min) (0 to 5 - hard level, 5 to 10 - medium level, 10 to 15 - easy level)')
plt.legend(loc=1)
plt.show()
```
